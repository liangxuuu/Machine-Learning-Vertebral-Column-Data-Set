# Dividing the Data into training and test sets as per b(iii)

```python
training_data = pd.concat([normal[0:70], abnormal[0:140]])
training_data.reset_index(drop=True,inplace=True)
# training_data = training_data.sample(frac=1).reset_index(drop=True)
training_data
```

```python
test_data = pd.concat([normal[70:], abnormal[140:]])
test_data.reset_index(drop=True,inplace=True)
# test_data = test_data.sample(frac=1).reset_index(drop=True)
test_data
```

```python
X_training = np.array(training_data[['pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle', 'sacral_slope', 'pelvic_radius', 'degree_spondylolisthesis']])
y_training = np.array(training_data['class'])
y_training = np.array([0 if u == 'Normal' else 1 for u in y_training])

X_test = np.array(test_data[['pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle', 'sacral_slope', 'pelvic_radius', 'degree_spondylolisthesis']])
y_test = np.array(test_data['class'])
y_test = np.array([0 if u == 'Normal' else 1 for u in y_test])
```

# Code for k-nearest neighbors with Euclidean metric  as per c(i)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error

train_error_throughtout = []
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_training, y_training)
```

```python
y_pred = knn.predict(X_test)
y_train_pred = knn.predict(X_training)
```

```python
from sklearn.metrics import accuracy_score
y_test_error = 1 - accuracy_score(y_test, y_pred)
y_train_error = 1 - accuracy_score(y_training, y_train_pred)
train_error_throughtout.append(y_train_error)
y_test_error
```

# Getting the optimal k with lowest error rate as per c(ii)

```python
k = np.arange(1,211,3)
training_error = {}
test_error_array = []
training_error_array = []
test_error = {}
k = k[::-1]
for i in k:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_training, y_training)
    pred_test = knn.predict(X_test)
    pred_training = knn.predict(X_training)
    test_error_array.append(1 - accuracy_score(y_test, pred_test))
    training_error_array.append(1 - accuracy_score(y_training, pred_training))
    test_error.update({i: 1 - accuracy_score(y_test, pred_test)})
    training_error.update({i: 1 - accuracy_score(y_training, pred_training)})
test_error
```

```python
test_error_array = np.array(test_error_array)
print(train_error_throughtout)
training_error_array = np.array(training_error_array)

train_error_throughtout.append(min(training_error_array))
```

# Plotting training error and test error vs the values of K as per c(ii)

```python
plt.plot(k, test_error_array, color='red', markerfacecolor='blue')
plt.plot(k, training_error_array, color='blue', markerfacecolor='blue')
plt.xlim(210,-5)
plt.xlabel("K")
plt.ylabel("Errors")
```

```python
optimal_k = [key for m in [min(test_error.values())] for key,val in test_error.items() if val == m][-1]
optimal_k
```

# Using subset of training data and getting the lowest error for different size of training data as per c(iii)

```python
best_test_error_rate = {}
best_train_error_rate = {}
test_error_array = []
train_error_array = []
for n in range(10, 211, 10):
    test_error_array = []
    train_error_array = []
    temp_training_data = pd.concat([training_data[0:int(round(n/3))], training_data[70:70+int(round(2*n/3))]])
    temp_training_data.reset_index(drop=True,inplace=True)
    
    X_temp_training_data = np.array(temp_training_data[['pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle', 'sacral_slope', 'pelvic_radius', 'degree_spondylolisthesis']])
    y_temp_training_data = np.array(temp_training_data['class'])
    y_temp_training_data = np.array([0 if u == 'Normal' else 1 for u in y_temp_training_data])
    
    for k in range(1, n, 5):
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_temp_training_data, y_temp_training_data)
        pred_test = knn.predict(X_test)
        test_error_array.append(1- accuracy_score(y_test, pred_test))
        pred_train = knn.predict(X_temp_training_data)
        train_error_array.append(1- accuracy_score(y_temp_training_data, pred_train))
    best_test_error_rate.update({n: min(test_error_array)})
    best_train_error_rate.update({n: min(train_error_array)})
best_test_error_rate
```

# Plotting the learning curve as per c(iii)

```python
train_error_throughtout.append(min(best_train_error_rate.values()))

plt.plot(range(10, 211, 10), best_test_error_rate.values(), color='blue', markerfacecolor='blue')
plt.xlabel("N")
plt.ylabel("Errors")
```

# Code for k-nearest neighbors with Mahalanobis Distance as per d(ii)

```python
k = np.arange(1,200,5)
test_error_array = {}
train_error_array = {}
covarience_x = np.cov(X_training)  
covarience_x_inverse = np.linalg.inv(covarience_x)
for i in k:
    knn = KNeighborsClassifier(n_neighbors=i,algorithm='brute', 
                               metric='mahalanobis', 
                               metric_params={'VI': covarience_x_inverse})
    knn.fit(X_training, y_training)
    y_test_pred = knn.predict(X_test)
    test_error_array.update({i: 1 - accuracy_score(y_test, y_test_pred)})
    y_train_pred = knn.predict(X_training)
    train_error_array.update({i: 1 - accuracy_score(y_training, y_train_pred)})
test_error_array
```

```python
train_error_throughtout.append(min(train_error_array.values()))
optimal_k = [key for m in [min(test_error_array.values())] for key,val in test_error_array.items() if val == m][0]
optimal_k
```
